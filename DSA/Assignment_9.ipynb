Given an integer¬†`n`, return¬†*`true`¬†if it is a power of two. Otherwise, return¬†`false`*.

An integer¬†`n`¬†is a power of two, if there exists an integer¬†`x`¬†such that¬†`n == 2x`.

**Example 1:**
Input: n = 1 

Output: true

**Example 2:**
Input: n = 16 

Output: true

**Example 3:**
Input: n = 3 

Output: false

</aside>

```python
def is_power_2(n):
    if n < 0 or n%2 != 0:
        return False
    if n%2 == 0:
        return True
    is_power_2(n%2)
    
is_power_2(8)
```

Given a number n, find the sum of the first natural numbers.

**Example 1:**

Input: n = 3 

Output: 6

**Example 2:**

Input  : 5 

Output : 15

</aside>

```python
def sum_n(n):
    sum1 = 0
    if n== 0:
        return 0
    if n>0:
        sum1 = n + sum_n(n-1)
    return sum1
sum_n(3)
```

****Given a positive integer, N. Find the factorial of N. 

**Example 1:**

Input: N = 5 

Output: 120

**Example 2:**

Input: N = 4

Output: 24

</aside>

```python
def fact(n):
    if n==1:
        return 1
    return n*fact(n-1)
fact(5)
```

Given a number N and a power P, the task is to find the exponent of this number raised to the given power, i.e. N^P.

**Example 1 :** 

Input: N = 5, P = 2

Output: 25

**Example 2 :**
Input: N = 2, P = 5

</aside>

```python
def expo(p,n):
    if n==0:
        return 1
    return p*expo(p,n-1)
expo(5,2)
```

Given an array of integers¬†**arr**, the task is to find maximum element of that array using recursion.

**Example 1:**

Input: arr = {1, 4, 3, -5, -4, 8, 6};
Output: 8

**Example 2:**

Input: arr = {1, 4, 45, 6, 10, -8};
Output: 45

</aside>

```python
def get_max(arr,i,value):
    if i != len(arr):
        if value < arr[i]:
            value = arr[i]
#             print(value)
        i += 1
        get_max(arr,i,value)
    else:
        print(value)

get_max([1, 4, 3, -5, -4, 8, 6],0,0)
```

üí° **Question 6**

Given first term (a), common difference (d) and a integer N of the Arithmetic Progression series, the task is to find Nth term of the series.

**Example 1:**

Input : a = 2 d = 1 N = 5
Output : 6
The 5th term of the series is : 6

**Example 2:**

Input : a = 5 d = 2 N = 10
Output : 23
The 10th term of the series is : 23

</aside>

```python
def AP(a,d,n):
    if n==1:
        return a
    if n != 0:
        a+=d
        return AP(a,d,n-1)
AP(5,2,10)
        
```

Given a string S, the task is to write a program to print all permutations of a given string.

**Example 1:**

***Input:***

*S = ‚ÄúABC‚Äù*

***Output:***

*‚ÄúABC‚Äù, ‚ÄúACB‚Äù, ‚ÄúBAC‚Äù, ‚ÄúBCA‚Äù, ‚ÄúCBA‚Äù, ‚ÄúCAB‚Äù*

**Example 2:**

***Input:***

*S = ‚ÄúXY‚Äù*

***Output:***

*‚ÄúXY‚Äù, ‚ÄúYX‚Äù*

</aside>

```python
def permutation_string(a,l):
    r=len(a)
    if l == r:
        print("".join(a), end=" ")
    else:
        for i in range(l, r):
            a[l], a[i] = a[i], a[l]
            permutation_string(a, l+1)
            a[l], a[i] = a[i], a[l] 

a=list("ABC")
permutation_string(a,0)
```

üí° **Question 8**

Given an array, find a product of all array elements.

**Example 1:**

Input  : arr[] = {1, 2, 3, 4, 5}
Output : 120
**Example 2:**

Input  : arr[] = {1, 6, 3}
Output : 18

</aside>

```python
def prodct(arr,n):
    if n==len(arr)-1:
        return arr[n]
    return arr[n]*prodct(arr,n+1)

prodct([1, 2, 3, 4, 5],0)
```
